```python
from bs4 import BeautifulSoup
import sys 
import requests
import pandas as pd

res = []
posts = 0

for i in range(404, 2000):
    # if i % 2 == 0:
    print("Page: ", i, "/2000")
    print("Post: ", posts, " -------------------------------------- \n\n\n\n")
    df = pd.DataFrame(res, columns=['content'])
    df.to_csv('/content/gdrive/MyDrive/datasets/0/vnexpress-chinhtri_2.csv')

    url = "https://vnexpress.net/thoi-su/chinh-tri-p" + str(i)
    page = requests.get(url)

    soup = BeautifulSoup(page.content, 'lxml')
    for foo in soup.find_all('h2', attrs={'class': 'title-news'}):
        foo_descendants = foo.descendants
        for d in foo_descendants:
            if d.name == 'a':
                link = d['href']
                
                article = requests.get(link)
                article_soup = BeautifulSoup(article.content, 'lxml')

                try:
                    title = article_soup.find('h1', {"class":"title-detail"}).text
                    description = article_soup.find('p', {"class":"description"}).text
                    content = article_soup.find('article', {"class":"fck_detail"}).text
                    temp = title + '\n' + description + '\n' + content
                    print("Title: ", title)
                    print(content[:200])
                    print("*****************************************")
                    res.append(temp)
                    posts += 1

                except:
                    print("Skip page: ", link)
```

```python
driver = webdriver.Chrome('chromedriver',chrome_options=chrome_options)
driver.get('https://www.luatkhoa.org/topics/xa-hoi/')

# SCROLL---------------------------------------------
SCROLL_PAUSE_TIME = 0.5

# # Get scroll height
# last_height = driver.execute_script("return document.body.scrollHeight")

# it = 0
# while it < 5000:
#     # Scroll down to bottom
#     driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")

#     # Wait to load page
#     time.sleep(SCROLL_PAUSE_TIME)
#     it += 1

#     # # Calculate new scroll height and compare with last scroll height
#     new_height = driver.execute_script("return document.body.scrollHeight")
#     if new_height == last_height:
#         break
#     last_height = new_height

while driver.find_element_by_css_selector('#mvp-main-body-wrap > div > div > div > div > div.mvp-main-blog-in > div > div > a'):
  load_more_btn = driver.find_element_by_css_selector('#mvp-main-body-wrap > div > div > div > div > div.mvp-main-blog-in > div > div > a')
  try:
    load_more_btn.click()
    time.sleep(4)
    cprint("Load more...", 'blue')
  except:
    cprint("Load more error", 'red')
    break


csv = []
posts = 0

# Get elements ---------------------------------------------
# copy selector from webpage
cards = driver.find_elements_by_css_selector("#mvp-main-body-wrap > div > div > div > div > div.mvp-main-blog-in > div > ul > li > a")
for i, card in enumerate(cards):
    try:
        # open new tab
        # pyautogui.hotkey('ctrl', 't')
        driver.execute_script('''window.open('',"_blank");''')

        # get link
        link = card.get_attribute('href')
        time.sleep(1)
        driver.switch_to.window(driver.window_handles[-1])
        driver.get(link)

        # Get content of article
        title = driver.find_element_by_css_selector('#mvp-post-head > h1').text
        content = driver.find_elements_by_css_selector('#mvp-content-main > p')
        
        temp = title + '\n'
        for para in content:
          temp += para.text + '. \n'

        print("Title: ", title)
        print("Content: ", temp[:100], '\n**************************')
        row = title + '\n' + temp

        csv.append(row)
        posts += 1


        # # go back
        # # pyautogui.hotkey('ctrl', 'w')
        driver.close()
        driver.switch_to.window(driver.window_handles[-1])

        time.sleep(1)

    except:
        cprint('Page not found', 'red')

    # # save file
    if i % 5 == 0:
        print_out = '\n\n-------------Scraping: {} / 10000  ----- Posts: {} -----------------------------\n\n'.format(i, posts)
        cprint(print_out, 'green', attrs= ['reverse'])
        df = pd.DataFrame(csv, columns=['blue'])
        df.to_csv('/content/gdrive/MyDrive/Internship/Phat hien nd phan dong/datasets/1/KhoaLuatTapChi_XaHoi.csv')


time.sleep(3)

driver.quit()
```

## Thống kê dataset

```python
import os
import pandas as pd
import numpy as np

path = r'/content/gdrive/MyDrive/Internship/Phat hien nd phan dong/datasets/0/'
os.chdir(path)

num_of_lines = 0
avg_num_words = []

for filename in os.listdir(path):
    df = pd.read_csv(filename)
    n_lines = df.shape[0]
    if "KhoaLuat" in filename:
        df = df.rename(columns={"blue": "content"})
    df['len_word'] = df['content'].str.split().str.len()


    print("Rows Counted {} in the csv {} with {:.0f} of single words:".format(n_lines, filename, np.round(np.mean(df['len_word']))))
    num_of_lines += n_lines
    avg_num_words.append(np.round(np.nanmedian(df['len_word'])))

print("Total: ", num_of_lines)

print(np.mean(avg_num_words))
```
